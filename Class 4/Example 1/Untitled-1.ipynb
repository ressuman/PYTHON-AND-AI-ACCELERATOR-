```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Data: x values from 1 to 10, y values as squares
x = np.arange(1, 11)       # array([1, 2, ..., 10])
y = x ** 2                 # squares of x

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', color='royalblue', linewidth=2)

# Labels and title
plt.xlabel("X Values")
plt.ylabel("Y = X Squared")
plt.title("Line Chart of X vs X Squared")

# Add grid
plt.grid(True)

# Show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data directly from GitHub
url = "https://raw.githubusercontent.com/azhar2ds/DataSets/master/10000%20Sales%20Records.csv"
df = pd.read_csv(url, parse_dates=["Order Date"])

# Filter for 2014 (dataset has 2014–2016)
df_2014 = df[df["Order Date"].dt.year == 2014]

# Aggregate monthly totals
monthly = (
    df_2014.groupby(df_2014["Order Date"].dt.month)["Total Revenue"].sum().reset_index()
)

# Plot
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(monthly["Order Date"], monthly["Total Revenue"], color="#E94560")
ax.set_title("Monthly Revenue 2014")
ax.set_xlabel("Month")
ax.set_ylabel("Revenue")

# Use only the months present in the data
month_labels = [
    "Jan",
    "Feb",
    "Mar",
    "Apr",
    "May",
    "Jun",
    "Jul",
    "Aug",
    "Sep",
    "Oct",
    "Nov",
    "Dec",
]
ax.set_xticks(monthly["Order Date"])
ax.set_xticklabels([month_labels[m - 1] for m in monthly["Order Date"]])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Create sample data (dates + random revenue)
dates = pd.date_range(start="2024-01-01", end="2024-12-31", freq="D")
np.random.seed(42)  # reproducible results
revenue = np.random.randint(1000, 5000, size=len(dates))

df = pd.DataFrame({"date": dates, "revenue": revenue})

# 2. Filter — only 2024 data
df_2024 = df[df["date"].dt.year == 2024]

# 3. Aggregate — monthly totals
monthly = df_2024.groupby(df_2024["date"].dt.month)["revenue"].sum().reset_index()

# 4. Visualise
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(monthly["date"], monthly["revenue"], color="#E94560")
ax.set_title("Monthly Revenue 2024")
ax.set_xlabel("Month")
ax.set_ylabel("Revenue")

# ✅ Correct month labels list
month_labels = [
    "Jan",
    "Feb",
    "Mar",
    "Apr",
    "May",
    "Jun",
    "Jul",
    "Aug",
    "Sep",
    "Oct",
    "Nov",
    "Dec",
]

# Match ticks to months present in data
ax.set_xticks(monthly["date"])
ax.set_xticklabels([month_labels[m - 1] for m in monthly["date"]])

plt.tight_layout()
plt.show()
```

```python
# LINE CHART
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 145, 162, 158, 190, 210]

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(
    months,
    sales,
    color="#E94560",
    linewidth=2.5,
    marker="o",
    markersize=7,
    label="Monthly Sales",
)

ax.fill_between(months, sales, alpha=0.1, color="#E94560")  # shaded area
ax.set_title("Sales Trend — H1", fontsize=14)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# BAR CHART (vertical & horizontal)
categories = ["Python", "R", "SQL", "Java", "Scala"]
counts = [85, 60, 72, 45, 30]
colors = ["#E94560", "#F5A623", "#0F3460", "#16213E", "#94A3B8"]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Vertical bars
axes[0].bar(categories, counts, color=colors, edgecolor="white")
axes[0].set_title("Vertical Bar")

# Horizontal bars (better for long labels)
axes[1].barh(categories, counts, color=colors)
axes[1].set_title("Horizontal Bar")

plt.tight_layout()
plt.show()
```

```python
# SCATTER PLOT
import numpy as np

np.random.seed(42)

x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.8
size = np.abs(np.random.randn(100)) * 100

fig, ax = plt.subplots(figsize=(7, 5))
scatter = ax.scatter(
    x,
    y,
    c=y,  # colour by y-value
    s=size,  # point size by 3rd variable
    cmap="RdYlBu",  # colour map
    alpha=0.7,
    edgecolors="white",
    linewidths=0.5,
)

plt.colorbar(scatter, ax=ax, label="Value")
ax.set_title("Relationship between X and Y")
plt.tight_layout()
plt.show()
```

```python
# HISTOGRAM
data = np.random.normal(loc=50, scale=15, size=500)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Basic histogram
axes[0].hist(data, bins=30, color="#E94560", edgecolor="white", alpha=0.85)
axes[0].set_title("Distribution of Scores")

# Overlapping histograms (comparing 2 groups)
data2 = np.random.normal(60, 12, 500)
axes[1].hist(data, bins=25, alpha=0.6, label="Group A", color="#E94560")
axes[1].hist(data2, bins=25, alpha=0.6, label="Group B", color="#0F3460")
axes[1].legend()
plt.tight_layout()
plt.show()
```

```python
# HEATMAP — correlation matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load sample dataset (Seaborn has built-in "flights")
df = sns.load_dataset("flights")

# 2. Pivot to wide format (months as rows, years as columns)
pivot = df.pivot(index="month", columns="year", values="passengers")

# 3. Plot heatmap
fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(
    pivot,
    annot=True,  # show numbers in cells
    fmt="d",  # integer format
    cmap="YlOrRd",
    linewidths=0.5,
    ax=ax,
)

ax.set_title("Passengers per Month & Year")
plt.tight_layout()
plt.show()
```

```python
# PAIRPLOT — all-vs-all scatter matrix
df = sns.load_dataset("iris")

# Basic pairplot
sns.pairplot(df, hue="species", palette="Set1")

# Customised pairplot
sns.pairplot(
    df,
    hue="species",
    diag_kind="kde",  # KDE on diagonal
    plot_kws={"alpha": 0.6, "s": 60, "edgecolor": "white"},
    palette="husl",
)

plt.suptitle("Iris Feature Pairs", y=1.02, fontsize=14)
plt.show()
```

```python
# BOX PLOT
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load sample dataset (Seaborn has built-in "tips")
df = sns.load_dataset("tips")

# 2. Box plot
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(data=df, x="day", y="total_bill", hue="smoker", palette="Set2", ax=ax)

# 3. Strip plot overlay
sns.stripplot(data=df, x="day", y="total_bill", color="black", alpha=0.3, size=3, ax=ax)

plt.title("Total Bill Distribution by Day and Smoker Status")
plt.show()
```
