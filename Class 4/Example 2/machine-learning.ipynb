```python
# Supervised learning example: predict y from x
from sklearn.linear_model import LinearRegression
import numpy as np

# Training data (x, y)
X = np.array([[1], [2], [3], [4]])  # study hours
y = np.array([2, 4, 6, 8])  # test scores

# Train the model
model = LinearRegression()
model.fit(X, y)

# Predict new value
print("If I study 5 hours, score will be:", model.predict([[5]]))
```

```python
# Unsupervised learning example: cluster data
from sklearn.cluster import KMeans
import numpy as np

# Data points (like animals with features)
X = np.array([[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0]])

# Group into 2 clusters
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)

print("Cluster labels:", kmeans.labels_)
print("Cluster centers:", kmeans.cluster_centers_)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans

# Supervised example
X = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])
model = LinearRegression().fit(X, y)
print("Study 5 hours → score:", model.predict([[5]]))

# Unsupervised example
data = np.array([[1, 2], [1, 4], [10, 2], [10, 4]])
kmeans = KMeans(n_clusters=2, random_state=0).fit(data)
print("Cluster labels:", kmeans.labels_)
```

```python
# ---------------------------------------------------------
# Temperature Conversion Visualization: Celsius ↔ Fahrenheit
# ---------------------------------------------------------

# 1. Import required libraries
import matplotlib.pyplot as plt
import numpy as np

# 2. Define known data points (freezing and boiling of water)
celsius = np.array([0, 100])  # Celsius values
fahrenheit = np.array([32, 212])  # Corresponding Fahrenheit values

# 3. Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# 4. Plot the known points as large teal dots
ax.scatter(celsius, fahrenheit, color="teal", s=200, zorder=5, label="Known Points")

# 5. Draw the straight line through those points
ax.plot(celsius, fahrenheit, color="coral", linewidth=3, label="Conversion Line")

# 6. Label the axes
ax.set_xlabel("Celsius (°C)", fontsize=12)
ax.set_ylabel("Fahrenheit (°F)", fontsize=12)

# 7. Add a descriptive title
ax.set_title(
    "Temperature Conversion: Linear Relationship", fontsize=14, fontweight="bold"
)

# 8. Add a light grid for readability
ax.grid(True, alpha=0.3)

# 9. Show legend
ax.legend(fontsize=11)

# 10. Annotate the freezing and boiling points
ax.annotate(
    f"({celsius[0]}°C, {fahrenheit[0]}°F)",
    xy=(celsius[0], fahrenheit[0]),
    xytext=(10, 10),
    textcoords="offset points",
    fontsize=10,
)

ax.annotate(
    f"({celsius[1]}°C, {fahrenheit[1]}°F)",
    xy=(celsius[1], fahrenheit[1]),
    xytext=(-80, -25),
    textcoords="offset points",
    fontsize=10,
)

# 11. Add the formula text on the plot (for teaching)
ax.text(10, 180, "F = 1.8C + 32", fontsize=12, color="purple")

# 12. Adjust layout to prevent overlap
plt.tight_layout()

# 13. Display the graph
plt.show()
```

```python
# ---------------------------------------------------------
# Study Hours vs Test Scores Visualization
# ---------------------------------------------------------

# 1. Import libraries
import matplotlib.pyplot as plt
import numpy as np

# 2. Define data (hours studied vs exam scores)
hours = np.array([1, 2, 3, 4, 5])
scores = np.array([50, 60, 65, 70, 80])

# 3. Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# 4. Plot the data points
ax.scatter(hours, scores, color="teal", s=200, label="Observed Scores")

# 5. Fit a simple line (y = m·x + c)
m, c = np.polyfit(hours, scores, 1)
ax.plot(hours, m * hours + c, color="coral", linewidth=3, label="Trend Line")

# 6. Label axes
ax.set_xlabel("Study Hours", fontsize=12)
ax.set_ylabel("Test Score (%)", fontsize=12)

# 7. Title
ax.set_title(
    "Relationship Between Study Hours and Test Scores", fontsize=14, fontweight="bold"
)

# 8. Grid + Legend
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)

# 9. Annotate a point
ax.annotate(
    "3 hrs → 65%", xy=(3, 65), xytext=(20, -30), textcoords="offset points", fontsize=10
)

# 10. Show formula
ax.text(1, 75, f"Score = {m:.1f}·Hours + {c:.1f}", fontsize=12, color="purple")

plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------------
# Monthly Sales Growth Visualization
# ---------------------------------------------------------

# 1. Import libraries
import matplotlib.pyplot as plt
import numpy as np

# 2. Define monthly sales data
months = np.arange(1, 7)  # Jan to Jun
sales = np.array([200, 240, 300, 350, 400, 480])

# 3. Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# 4. Plot sales as bars
ax.bar(months, sales, color="skyblue", label="Monthly Sales")

# 5. Fit a trend line
m, c = np.polyfit(months, sales, 1)
ax.plot(months, m * months + c, color="coral", linewidth=3, label="Growth Trend")

# 6. Label axes
ax.set_xlabel("Month", fontsize=12)
ax.set_ylabel("Sales (Units)", fontsize=12)

# 7. Title
ax.set_title("Sales Growth Over 6 Months", fontsize=14, fontweight="bold")

# 8. Grid + Legend
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)

# 9. Annotate peak
ax.annotate(
    "Highest: 480 units",
    xy=(6, 480),
    xytext=(-80, -30),
    textcoords="offset points",
    fontsize=10,
)

plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------------
# Height vs Weight Visualization
# ---------------------------------------------------------

# 1. Import libraries
import matplotlib.pyplot as plt
import numpy as np

# 2. Define sample data
height = np.array([150, 160, 170, 180, 190])  # cm
weight = np.array([50, 55, 65, 75, 85])  # kg

# 3. Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# 4. Plot data points
ax.scatter(height, weight, color="teal", s=200, label="Individuals")

# 5. Fit regression line
m, c = np.polyfit(height, weight, 1)
ax.plot(height, m * height + c, color="coral", linewidth=3, label="Trend Line")

# 6. Label axes
ax.set_xlabel("Height (cm)", fontsize=12)
ax.set_ylabel("Weight (kg)", fontsize=12)

# 7. Title
ax.set_title("Relationship Between Height and Weight", fontsize=14, fontweight="bold")

# 8. Grid + Legend
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)

# 9. Annotate a point
ax.annotate(
    "170 cm → 65 kg",
    xy=(170, 65),
    xytext=(-80, 30),
    textcoords="offset points",
    fontsize=10,
)

# 10. Show formula
ax.text(150, 80, f"Weight = {m:.2f}·Height + {c:.1f}", fontsize=12, color="purple")

plt.tight_layout()
plt.show()
```
