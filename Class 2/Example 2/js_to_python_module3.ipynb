## 1 — Null Checks & Defaults

```python
# JS:  if (user === null || user === undefined) { console.log("no user") }
# NEW: Python uses 'is None' for identity checks

user = None

if user is None:
    print("no user")

# Shorthand — None, 0, "", [] are all falsy in Python
if not user:
    print("still no user (shorthand)")
```

```python
# JS:  const name = user.name || "Guest"
# NEW: use 'or' as the fallback operator

user = {"email": "ama@example.com"}  # no 'name' key

name = user.get("name") or "Guest"
print(name)  # Guest

# Walrus operator (Python 3.8+) — assign and check in one line
# JS: if (name = user.name) { ... }  <-- this is actually bad in JS but walrus is clean in Python
if email := user.get("email"):
    print(f"Found email: {email}")
```

## 2 — Loops

```python
# JS:  for (const lang of langs) { console.log(lang) }
# NEW: for...of becomes 'for x in list'

langs = ["JS", "Python", "Go"]

for lang in langs:
    print(lang)
```

```python
# JS:  for (const [i, lang] of langs.entries()) { console.log(i, lang) }
# NEW: .entries() becomes enumerate()

for i, lang in enumerate(langs):
    print(i, lang)
```

```python
# JS:  for (const [key, val] of Object.entries(user)) { console.log(key, val) }
# NEW: Object.entries() becomes .items() on a dict

user = {"name": "Ama", "age": 25, "role": "AI Engineer"}

for key, val in user.items():
    print(f"{key}: {val}")
```

```python
# JS:  for (let i = 0; i < 5; i++) { console.log(i) }
# NEW: classic for-loop becomes range()

for i in range(5):
    print(i)

print("---")

# range(start, stop, step) — count by 2s
for i in range(0, 10, 2):
    print(i)  # 0 2 4 6 8
```

## 3 — Array / List Higher-Order Methods

```python
# JS:  const doubled = nums.map(n => n * 2)
# NEW: .map() becomes a list comprehension
# Pattern: [expression for item in iterable]

nums = [1, 2, 3, 4]

doubled = [n * 2 for n in nums]
print(doubled)  # [2, 4, 6, 8]
```

```python
# JS:  const evens = nums.filter(n => n % 2 === 0)
# NEW: .filter() becomes a list comprehension with 'if'
# Pattern: [expression for item in iterable if condition]

evens = [n for n in nums if n % 2 == 0]
print(evens)  # [2, 4]
```

```python
# JS:  const result = nums.filter(n => n > 2).map(n => n * 10)
# NEW: filter + map chain becomes one comprehension

result = [n * 10 for n in nums if n > 2]
print(result)  # [30, 40]
```

```python
# JS:  const total = nums.reduce((acc, n) => acc + n, 0)
# NEW: Python has built-ins for common reduce operations

total    = sum(nums)   # 10
biggest  = max(nums)   # 4
smallest = min(nums)   # 1

print(f"sum={total}, max={biggest}, min={smallest}")

# For custom reduce logic, use functools.reduce
total_custom = reduce(lambda acc, n: acc + n, nums)
print(f"custom reduce: {total_custom}")
```

```python
# JS:  const found = users.find(u => u.id === 3)  // returns first match or undefined
# NEW: use next() with a generator expression

users = [
    {"id": 1, "name": "Kwame"},
    {"id": 2, "name": "Ama"},
    {"id": 3, "name": "Kofi"},
]

found = next((u for u in users if u["id"] == 3), None)
print(found)  # {'id': 3, 'name': 'Kofi'}

not_found = next((u for u in users if u["id"] == 99), None)
print(not_found)  # None
```

## 4 — Functions

```python
# JS:  function greet(name, greeting = "Hello") { return `${greeting}, ${name}!` }
# NEW: 'function' becomes 'def', template literals become f-strings

def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Kwame"))           # Hello, Kwame!
print(greet("Ama", "Welcome"))  # Welcome, Ama!
```

```python
# JS:  const square = n => n ** 2
# JS:  const add = (a, b) => a + b
# NEW: arrow functions become lambda (one-liners only)

square = lambda n: n ** 2
add    = lambda a, b: a + b

print(square(5))   # 25
print(add(3, 4))   # 7

# Lambdas shine as inline arguments (e.g. sorted, map)
names = ["Kofi", "Ama", "Kwame", "Akua"]
sorted_names = sorted(names, key=lambda n: len(n))
print(sorted_names)  # sorted by length
```

```python
# JS:  function logAll(label, ...items) { console.log(label, items) }
# NEW: rest params (...args) become *args

def log_all(label, *items):
    print(label, items)

log_all("Languages:", "Python", "Go", "Rust")
# Languages: ('Python', 'Go', 'Rust')

print("---")

# JS:  function setup({ host, port = 3000 }) { ... }
# NEW: named/object params become **kwargs

def setup(host, port=3000, **options):
    print(f"host={host}, port={port}, extras={options}")

setup("localhost", port=8000, debug=True, reload=True)
# host=localhost, port=8000, extras={'debug': True, 'reload': True}
```

## 5 — Destructuring & Spread

```python
# JS:  const [first, second, ...rest] = [1, 2, 3, 4, 5]
# NEW: Python uses the same star (*) syntax for unpacking

first, second, *rest = [1, 2, 3, 4, 5]

print(f"first={first}, second={second}, rest={rest}")
# first=1, second=2, rest=[3, 4, 5]

# Star can go in the middle too
*head, last = [10, 20, 30, 40]
print(f"head={head}, last={last}")
# head=[10, 20, 30], last=40
```

```python
# JS:  const { name, age, role = "viewer" } = user
# NEW: Python has no shorthand — unpack manually with .get()

user = {"name": "Ama", "age": 25}

name = user["name"]
age  = user["age"]
role = user.get("role", "viewer")  # default fallback like JS default params

print(f"{name}, {age}, {role}")
# Ama, 25, viewer
```

```python
# JS:  const merged = [...a, ...b]
# NEW: spread becomes * for lists

a = [1, 2]
b = [3, 4]
merged = [*a, *b]
print(merged)  # [1, 2, 3, 4]

print("---")

# JS:  const obj2 = { ...obj1, y: 2 }
# NEW: spread becomes ** for dicts

dict1 = {"x": 1}
dict2 = {**dict1, "y": 2}
print(dict2)  # {'x': 1, 'y': 2}

# Merging two dicts
defaults = {"theme": "light", "lang": "en", "debug": False}
overrides = {"theme": "dark", "debug": True}
config = {**defaults, **overrides}  # overrides wins on conflict
print(config)
```

## 6 — Classes

```python
# JS:  class Developer { constructor(name, lang) { this.name = name ... } }
# NEW: constructor -> __init__,  this -> self

class Developer:
    def __init__(self, name: str, lang: str):
        self.name = name
        self.lang = lang

    def introduce(self):
        return f"I'm {self.name}, I use {self.lang}"

    # __repr__ is like .toString() in JS
    def __repr__(self):
        return f"Developer(name={self.name!r}, lang={self.lang!r})"


# JS:  class SeniorDev extends Developer { ... }
# NEW: inheritance uses parentheses

class SeniorDev(Developer):
    def __init__(self, name: str, lang: str, years: int):
        super().__init__(name, lang)  # JS: super(name, lang)
        self.years = years

    def introduce(self):
        base = super().introduce()
        return f"{base} with {self.years} years experience"


dev = Developer("Kwame", "Python")
print(dev.introduce())

senior = SeniorDev("Ama", "Python", 5)
print(senior.introduce())
print(repr(senior))
```

```python
import httpx

# JS:  async function fetchUser(id) {
#        const res = await fetch(`/api/users/${id}`)
#        return await res.json()
#      }
# NEW: same structure, but use httpx instead of fetch

async def fetch_user(user_id: int):
    try:
        async with httpx.AsyncClient() as client:
            res = await client.get(f"https://jsonplaceholder.typicode.com/users/{user_id}")
            return res.json()
    except Exception as e:
        print(f"Error: {e}")
        return None

# In Jupyter, await works directly in a cell (no asyncio.run needed)
data = await fetch_user(1)
print(data["name"])    # Leanne Graham
print(data["email"])   # Sincere@april.biz
```

## 8 — Error Handling

```python
# JS:  try { JSON.parse(bad) } catch (err) { console.error(err.message) } finally { ... }
# NEW: catch -> except, and you can catch SPECIFIC exception types

bad_string = "{ not valid json }"

try:
    data = json.loads(bad_string)
except json.JSONDecodeError as e:
    print(f"JSON error: {e}")
except ValueError as e:
    print(f"Value error: {e}")
except Exception as e:
    # Catch-all — like catch(err) in JS
    print(f"Unknown error: {e}")
finally:
    # Always runs — same as JS finally
    print("Done (finally block ran)")
```

```python
# JS:  throw new Error("Something went wrong")
# NEW: raise Exception("...")

def divide(a, b):
    if b == 0:
        raise ValueError("Cannot divide by zero")  # JS: throw new Error(...)
    return a / b

try:
    result = divide(10, 0)
except ValueError as e:
    print(f"Caught: {e}")
```

## 9 — Bonus: Dictionary Comprehensions

> Like list comprehensions, but for building dicts. No direct JS equivalent.

```python
# JS:  Object.fromEntries(langs.map(lang => [lang, len(lang)]))
# NEW: dict comprehension — {key: value for item in iterable}

langs = ["Python", "Go", "Rust", "JavaScript"]

lang_lengths = {lang: len(lang) for lang in langs}
print(lang_lengths)
# {'Python': 6, 'Go': 2, 'Rust': 4, 'JavaScript': 10}

# With a condition — only langs with more than 3 characters
long_langs = {lang: len(lang) for lang in langs if len(lang) > 3}
print(long_langs)
```

## 10 — Bonus: Useful Built-ins with No JS Equivalent

```python
# zip() — loop over two lists at the same time
# JS: you'd use a for loop with index to access both arrays

names  = ["Kwame", "Ama", "Kofi"]
scores = [88, 95, 72]

for name, score in zip(names, scores):
    print(f"{name}: {score}")

print("---")

# Build a dict from two lists using zip
grade_book = dict(zip(names, scores))
print(grade_book)
# {'Kwame': 88, 'Ama': 95, 'Kofi': 72}
```

```python
# any() and all() — JS: .some() and .every()

scores = [88, 95, 72, 60, 45]

# JS:  scores.some(s => s >= 90)
has_high_score = any(s >= 90 for s in scores)
print(f"Any score >= 90? {has_high_score}")  # True

# JS:  scores.every(s => s >= 50)
all_passing = all(s >= 50 for s in scores)
print(f"All passing? {all_passing}")  # False
```

```python
# sorted() — JS: .sort() but non-destructive (returns a NEW list)

devs = [
    {"name": "Kofi",  "exp": 3},
    {"name": "Ama",   "exp": 7},
    {"name": "Kwame", "exp": 1},
]

# JS:  devs.sort((a, b) => a.exp - b.exp)
by_exp = sorted(devs, key=lambda d: d["exp"])

for d in by_exp:
    print(d["name"], d["exp"])

print("---")

# Reverse sort
by_exp_desc = sorted(devs, key=lambda d: d["exp"], reverse=True)
print([d["name"] for d in by_exp_desc])
```

## Quick Reference Table

| Concept | JavaScript | Python |
|---|---|---|
| Null check | `=== null` | `is None` |

| Some / Every | `.some()` / `.every()` | `any()` / `all()` |
| Sort | `.sort((a,b) => ...)` | `sorted(list, key=lambda x: ...)` |
