```python
import numpy as np

# 1. Create a simple array
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

# 2. Basic math operations
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))

# 3. Reshape into a matrix
matrix = arr.reshape(5, 1)
print("Reshaped Matrix:\n", matrix)

# 4. Random numbers
rand_nums = np.random.randint(0, 10, size=(3, 3))
print("Random 3x3 Matrix:\n", rand_nums)
```

```python
import pandas as pd

# 1. Create a DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40],
    "Score": [85, 90, 95, 88],
}
df = pd.DataFrame(data)
print("DataFrame:\n", df)

# 2. Basic statistics
print("Average Age:", df["Age"].mean())
print("Max Score:", df["Score"].max())

# 3. Filtering
print("People older than 30:\n", df[df["Age"] > 30])

# 4. Sorting
print("Sorted by Score:\n", df.sort_values(by="Score", ascending=False))
```

```python
import numpy as np
import pandas as pd

# Generate a 5x3 matrix of random values
values = np.random.rand(5, 3)

# Create DataFrame with custom row and column labels
df = pd.DataFrame(values, columns=["A", "B", "C"], index=["r1", "r2", "r3", "r4", "r5"])

print(df)
```
