# 1. Conditionals: The Syntax Switch

You already know logic from JavaScript. You just need the new keywords.

**Key Differences:** - JavaScript: `else if` and `{ }` - Python: `elif`
and **indentation**

````python
status = 404

if status == 200:
    print("Success")
elif status == 404:
    print("Not Found")  # Indentation matters!
else:
    print("Server Error")
````

# 2. The “Pythonic” Loop

**Golden Rule:** Developers often try to write C-style loops.
**Don’t.**  
In Python, we iterate directly over objects.

## A. Basic Iteration (`for...in`)

This is equivalent to JavaScript’s `for...of` loop.

````python
stack = [
    "React", "Express", "Node", "Python", 
    "JavaScript", "Java", "C++", "Rust", 
    "Zig", "Typescript", "OCaml", "R", 
    "Angular", "php", "Vue JS", "Lua", 
    "Odin", "Go", "Dart"
]

# The Python Way
for tech in stack:
    print(f"Learning: {tech}")
````

**Comparison:**

``` javascript
// JavaScript
for (const tech of stack) {
    console.log(`Learning: ${tech}`);
}
```

## B. The Index Problem (`enumerate`)

*“But how do I get the index `i`?”*  
Use `enumerate()`. It gives you a tuple of `(index, item)` each step.

````python
# JS Equivalent: stack.forEach((tech, i) => { ... })

for i, tech in enumerate(stack):
    print(f"Index {i}: {tech}")
````

> **Why enumerate()?**
>
> It’s cleaner and more Pythonic than using `range(len(stack))`. You get
> both the index and value automatically.

## C. Looping Two Lists (`zip`)

AI data often comes in separate lists (e.g., `images` and `labels`).  
Use `zip()` to loop them together like a zipper on a jacket.

````python
names = ["Kwame", "Ama", "Yaw"]
scores = [85, 92, 78]

# Loop simultaneously
for name, score in zip(names, scores):
    print(f"{name} scored {score}")
````

You are not limited to just two lists! You can zip three, four, or as
many iterables as you need.

````python
names = ["Alice", "Bob", "Charlie"]
scores = [85, 92, 78]
roles = ["Admin", "User", "Moderator"]

# Looping through three lists simultaneously
for name, score, role in zip(names, scores, roles):
    print(f"{name} ({role}) scored {score}")
````

### Critical Edge Cases with `zip()`

> **What happens with unequal lengths?**
>
> **`zip()` stops at the shortest list!** Data is silently lost.
>
> ``` python
> names = ["Kwame", "Ama", "Yaw", "Akosua"]  # 4 items
> scores = [85, 92]                           # 2 items
>
> for name, score in zip(names, scores):
>     print(f"{name}: {score}")
>
> # Output: Only Kwame and Ama!
> # Yaw and Akosua are IGNORED
> ```
>
>     Kwame: 85
>     Ama: 92
>
> **Safe Alternative:** Use `zip_longest()` from `itertools` if you need
> all items:
>
> ``` python
> from itertools import zip_longest
>
> names = ["Kwame", "Ama", "Yaw", "Akosua"]
> scores = [85, 92]
>
> for name, score in zip_longest(names, scores, fillvalue=0):
>     print(f"{name}: {score}")
>
> # Output includes all names, missing scores filled with 0
> ```
>
>     Kwame: 85
>     Ama: 92
>     Yaw: 0
>     Akosua: 0

### What can you `zip()`?

**Answer:** Any iterable! Not just lists.

````python
# Works with tuples
tuple1 = ("React", "Vue", "Angular")
tuple2 = ("Frontend", "Frontend", "Frontend")

for framework, category in zip(tuple1, tuple2):
    print(f"{framework} is a {category} framework")
````

````python
# Works with sets (but order is unpredictable!)
set1 = {"Python", "JavaScript", "Go"}
set2 = {"Backend", "Fullstack", "Backend"}

for lang, role in zip(set1, set2):
    print(f"{lang}: {role}")
# ⚠️ Order is NOT guaranteed with sets!
````

````python
# Even works with strings (they're iterables too)
letters = "ABC"
numbers = [1, 2, 3]

for letter, num in zip(letters, numbers):
    print(f"{letter}{num}")
````

> **Best Practice**
>
> -   Use `zip()` when you’re **certain** both iterables are the same
>     length
> -   Use `zip_longest()` when lengths might differ and you need all
>     data
> -   **Never** use `zip()` with sets in production code (unpredictable
>     order)
> -   Always validate data lengths before zipping in real applications

# 3. List Comprehensions (The “Killer Feature”)

This is the **most important tool** for data cleaning. It replaces
`.map()` and `.filter()`.

**Syntax:** `[expression for item in list if condition]`

## The Bridge: Traditional Loop vs. Comprehension

Before looking at the syntax, let’s look at *why* we use comprehensions.
Suppose we want to multiply every number in a list by 5.

Here is the traditional way using a standard `for` loop and `.append()`:

````python
nu_list = [2, 3, 7, 11, 3, 0.312, 45, 78, 32, 12, 71]

# 1. Create an empty list
prod = []

# 2. Loop through the original list
for item in nu_list:
    # 3. Do the math
    ans = item * 5
    # 4. Append to the new list
    prod.append(ans)

print("Traditional Loop:", prod)
````

Now, here is the exact same logic written as a **List Comprehension**.
It collapses those 4 lines of code into a single, elegant line:

````python
# [expression FOR item IN iterable]
prod2 = [item * 5 for item in nu_list]

print("List Comprehension:", prod2)
````

## Example 1: The `.map()` Equivalent

Multiply every number by 2.

````python
nums = [1, 2, 3, 4]

# JS: const doubled = nums.map(x => x * 2)
doubled = [x * 2 for x in nums]

print(doubled)
````

## Example 2: The `.filter()` Equivalent

Keep only even numbers.

````python
# JS: const evens = nums.filter(x => x % 2 === 0)
evens = [x for x in nums if x % 2 == 0]

print(evens)
````

## Example 3: Chaining (Filter + Map)

In JavaScript: `nums.filter(x => x % 2 === 0).map(x => x * 2)`

````python
# Filter AND transform in one line
result = [x * 2 for x in nums if x % 2 == 0]

print(result)
````

> **Real-World AI Connection: Feature Scaling**
>
> Before feeding data into an AI model, we often need to “scale” it
> (e.g., converting all numbers to be between 0 and 1). List
> comprehensions are the fastest way to apply a mathematical formula to
> thousands of data points instantly.

> **Readability First**
>
> If your list comprehension becomes too complex (nested conditions,
> multiple transformations), break it into a regular loop or multiple
> steps. Readability \> cleverness.

## From For Loop to List Comprehension: Walking Backwards

The key skill is **reverse engineering** a for loop into a
comprehension. Let’s learn the pattern.

### The Translation Pattern

Every list comprehension follows this structure:

``` python
result = [WHAT_TO_KEEP for ITEM in SOURCE if CONDITION]
```

**Step-by-step translation:**

1.  Find what’s being **appended** → This becomes `WHAT_TO_KEEP`
2.  Find the **loop variable** → This becomes `ITEM in SOURCE`
3.  Find any **if statement** → This becomes the `if CONDITION`

### Worked Example: Temperature Converter

Let’s convert this loop together:

````python
# Original For Loop
celsius_temps = [0, 10, 20, 30, 40]
fahrenheit = []

for temp in celsius_temps:
    if temp >= 20:  # Only convert warm temperatures
        converted = temp * 9/5 + 32
        fahrenheit.append(converted)

print(fahrenheit)
````

**Translation Steps:**

``` python
# Step 1: What's being appended?
#         → converted = temp * 9/5 + 32

# Step 2: What's the loop?
#         → for temp in celsius_temps

# Step 3: What's the condition?
#         → if temp >= 20

# Combine them:
fahrenheit = [temp * 9/5 + 32 for temp in celsius_temps if temp >= 20]
```

Let’s verify they produce the same result:

````python
# The comprehension version
celsius_temps = [0, 10, 20, 30, 40]
fahrenheit = [temp * 9/5 + 32 for temp in celsius_temps if temp >= 20]

print(fahrenheit)  # [68.0, 86.0, 104.0]
````

> **Reading Comprehensions Aloud**
>
> Try reading it in English: “Give me `temp * 9/5 + 32` for each `temp`
> in `celsius_temps` if `temp >= 20`”

### Example Exercise: Student Grades

Let’s do another one together. Here’s the for loop:

````python
# Original For Loop
students = ["Kwame", "Ama", "Yaw", "Akosua", "Kofi"]
grades = [45, 78, 52, 88, 91]
passing_students = []

for i, student in enumerate(students):
    if grades[i] >= 50:
        passing_students.append(student)

print(passing_students)
````

**How to convert this:**

``` python
# Step 1: What's appended? → student
# Step 2: Loop structure? → We need BOTH student and grade
#         Solution: use zip() to pair them!
# Step 3: Condition? → if grade >= 50

# The comprehension:
passing_students = [student for student, grade in zip(students, grades) if grade >= 50]
```

````python
# Verify it works
students = ["Kwame", "Ama", "Yaw", "Akosua", "Kofi"]
grades = [45, 78, 52, 88, 91]

passing_students = [student for student, grade in zip(students, grades) if grade >= 50]
print(passing_students)  # ['Ama', 'Yaw', 'Akosua', 'Kofi']
````

# 4. Lambda Functions: A Brief Intro

In JavaScript, you often use arrow functions for quick operations:
`(x) => x * 2`. Python’s equivalent is the **lambda function**.

We use lambdas for **quick, one-line operations** where defining a full
function with `def` would be overkill. You’ll see these most often when
sorting data or using specific libraries.

``` python
# The Syntax: lambda arguments: expression
double = lambda x: x * 2
print(double(5)) # 10
```

### Common Use: Sorting Data

You have a list of students and want to sort by age. The `key` parameter
expects a function that tells it *how* to sort.

````python
students = [
    {"name": "Kwame", "age": 22},
    {"name": "Ama", "age": 19},
    {"name": "Yaw", "age": 25}
]

# Use lambda to tell sort() to look at the "age" key
students.sort(key=lambda s: s["age"])
print(students)
````

> **Moving Forward**
>
> We’ll dive much deeper into advanced Lambda patterns (like filtering,
> data mapping, and multi-level sorting) in **Lesson 3**.

## Real-World Example: Processing API Data

Here’s how lambda functions are used in actual data engineering:

````python
# You fetch weather data from an API
weather_data = [
    {"city": "Accra", "temp": 32, "humidity": 75},
    {"city": "Kumasi", "temp": 28, "humidity": 80},
    {"city": "Tamale", "temp": 35, "humidity": 60},
    {"city": "Cape Coast", "temp": 30, "humidity": 85}
]

# Find the hottest city
hottest = max(weather_data, key=lambda city: city["temp"])
print(f"Hottest city: {hottest['city']} at {hottest['temp']}°C")

# Find cities with high humidity (>70%)
humid_cities = list(filter(lambda c: c["humidity"] > 70, weather_data))
print(f"Humid cities: {[c['city'] for c in humid_cities]}")

# Extract just the temperatures
temps = list(map(lambda c: c["temp"], weather_data))
print(f"All temperatures: {temps}")
````

> **The “Pythonic” Way**
>
> While the above code works, experienced Python developers would write:
>
> ``` python
> # More readable and Pythonic
> hottest = max(weather_data, key=lambda c: c["temp"])  # Lambda is good here
> humid_cities = [c for c in weather_data if c["humidity"] > 70]  # Comprehension
> temps = [c["temp"] for c in weather_data]  # Comprehension
>
> # Rule of thumb: Use lambda for key functions, comprehensions for transformations
> ```

We use `def` to define functions. But for AI work, you need to
understand **Dynamic Arguments**.

## A. Default Arguments

You can set default values directly in the function definition.

````python
def greet(name, greeting="Akwaaba"):
    return f"{greeting}, {name}!"

print(greet("Kofi"))            # Uses default
print(greet("Kofi", "Hello"))   # Overrides default
````

## B. `*args` and `**kwargs` (The “Rest” Parameters)

AI libraries (like Transformers) use these to accept endless
configuration settings.

-   `*args`: Collects extra positional arguments into a **Tuple** (like
    JS `...args`)
-   `**kwargs`: Collects extra keyword arguments into a **Dictionary**

````python
def build_model(name, **config):
    print(f"Building: {name}")
    print("Config:", config)  # It's just a dict!

# We can pass ANY arguments we want
build_model("ChatBot", layers=12, speed="fast", gpu=True)
````

> **Real-World AI Example: The “Config Builder”**
>
> In AI libraries like **HuggingFace Transformers**, models take dozens
> of settings (dropout rates, attention heads, etc.). Instead of writing
> a function with 50 parameters, developers use `**kwargs`.
>
> ``` python
> # A realistic look at how AI models are initialized
> def initialize_ai_model(model_name, **hyperparameters):
>     print(f"Loading weights for: {model_name}")
>     for param, value in hyperparameters.items():
>         print(f"  -> Setting {param} to {value}")
>
> # The user can pass only the settings they want to change
> initialize_ai_model("BERT-Base", learning_rate=2e-5, batch_size=32, layers=12)
> ```

**Why this matters:** When you use HuggingFace or PyTorch, you’ll
constantly pass configuration objects this way.

````python
def super_function(*args, **kwargs):
    print("Positional arguments (tuple):", args)
    print("Keyword arguments (dict):", kwargs)

super_function(1, 2, 3, mode="turbo", verbose=True)
````

# 6. Modules: Organizing Code

In JavaScript, you use `export` and `import` (or `module.exports`).  
In Python, **every file is automatically a module**.

## A. Importing

Assume we have a file named `math_utils.py`:

``` python
# Import the whole file
import math_utils
math_utils.calculate()

# Import specific function (Like destructuring in JS)
from math_utils import calculate
calculate()

# Alias (Renaming) - Very common in data science
import pandas as pd
import numpy as np
```

## B. The `if __name__ == "__main__":` Block

You will see this everywhere. It’s a guard block.

Code inside this block **only runs if you execute the file directly**
(e.g., `python file.py`). It does **not** run if you import the file
into another script.

``` python
def main():
    print("Running main logic...")

if __name__ == "__main__":
    main()
```

**Use case:** You can write reusable functions at the top of a file, and
testing code at the bottom that only runs when you test the file
