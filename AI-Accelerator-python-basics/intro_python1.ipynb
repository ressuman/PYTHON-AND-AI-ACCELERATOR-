## A. Lists (The “Array”)

**Mutable, ordered sequences.** “Mutable” means you can change the list
after creating it - you can add items, remove items, or modify existing
items. Think of it like a shopping list you can edit anytime.

````python
# Creating
stack = ["React", "Express", "Node"]

# Accessing (Zero-indexed)
print(stack[0])  # React
print(stack[-1]) # Node (Negative indexing! JS can't do this natively)

# Slicing (The Power Move)
# [start:stop:step]
print(stack[0:2]) # ['React', 'Express']

# Adding
stack.append("Python") # JS: .push()
print(stack)
````

### Advanced Lists & Mixed Data Types

Lists in Python can hold mixed data types, including numbers, strings,
and even formatted f-strings. Let’s look at a more complex list and use
`len()` to check how many items it holds.

````python
# A list containing different types of string formats
tech_stack = [
    "Python",           # Single word
    "Data Science",     # String with a space
    "Agentic AI",       
    "Tamale",           # A city in Ghana
    "Machine Learning", 
    "Pandas",           
    "Jupyter Lab",      
    "uv package manager", 
    "2026",             # Number as a string
    f"User_{42}"        # An f-string evaluated dynamically
]

# Verify how many items are in our stack
print(f"Total items in tech_stack: {len(tech_stack)}")
````

You can also use a third parameter in slicing called the **step**. A
negative step lets you traverse the list backwards or skip elements.

````python
# [start:stop:step] - A step of -2 reads backwards, skipping every other item
print(tech_stack[::-2])
````

### Removing Items: The `.pop()` Method

While `.append()` adds to the end of a list, `.pop()` removes the last
item and *returns* it to you so you can use it in a variable.

````python
languages = ["Python", "JavaScript", "Java", "C++", "Rust", "Zig"]

# Remove and capture the last element
last_born = languages.pop()

print(f"Removed item: {last_born}")
print(f"Remaining list: {languages}")
````

> **Real-World AI Connection: Sequences**
>
> In AI, we use **Lists** to store sequences of data. For example, in
> NLP (Natural Language Processing), a sentence is often broken down
> into a list of words or “tokens”:
> `["The", "quick", "brown", "fox", "jumps", "over"]`.

### Understanding Indexing in Python

**Indexing** is how we access individual items in a list by their
position. Python uses **zero-based indexing**, meaning the first item is
at position 0, the second at position 1, and so on 😕.

````python
languages = ["Python", "JavaScript", "Java", "C++"]
#            [0]       [1]           [2]     [3]     <- Positive indices
#            [-4]      [-3]          [-2]    [-1]    <- Negative indices

# Examples:
print("First item (index 0):", languages[0])
print("Last item (index -1):", languages[-1])
print("Second to last (index -2):", languages[-2])
````

**Negative Indexing** is Python’s special feature that lets you count
from the end of the list backwards:

-   `-1` gives you the last item
-   `-2` gives you the second-to-last item
-   `-3` gives you the third-to-last item, and so on

**Why use negative indexing?** It’s incredibly useful when you don’t
know the length of a list but want to grab items from the end. In
JavaScript, you’d have to write `array[array.length - 1]` to get the
last item. In Python, you just write `array[-1]`. Much cleaner!

## B. Tuples (The “Constant Array”)

**New Concept:** Think of a Tuple as a List that is **read-only
(immutable)**.

**What does immutable mean?** Once you create a tuple, you **cannot
change its content**. You can’t add items, remove items, or modify
existing items. If you need to make changes, you must delete the old
tuple and create a completely new one.

-   **Why use it?** It’s faster than a list and protects data integrity.
    AI models often return data as tuples.
-   **Syntax:** Uses `(parentheses)` instead of `[brackets]`.

````python
# Defining a tuple
coordinates = (5.55602, -0.19690) # Accra Lat/Long
print(coordinates)
print(f"Latitude: {coordinates[0]}, Longitude: {coordinates[1]}")

# Uncommenting the line below would cause an error:
# coordinates[0] = 10.5 # ❌ ERROR! TypeError: 'tuple' object does not support item assignment

# If you need to change it, you must create a new tuple:
new_coordinates = (6.67, -1.57)  # New tuple with Kumasi coordinates
print("New coordinates:", new_coordinates)
````

> **Real-World AI Connection: Fixed Dimensions**
>
> In Computer Vision, we use **Tuples** for data that should never
> change during processing, like **image dimensions**
> `(height, width, channels)` or **bounding box coordinates**
> `(x_min, y_min, x_max, y_max)`. It prevents accidental bugs where code
> might try to “resize” the data incorrectly!

## C. Dictionaries (The “Object”)

**Key-Value pairs.** Similar to JSON/JS Objects, but Keys must be
**hashable**.

**What does “hashable” mean?** A hashable value is one that never
changes during its lifetime and can be compared to other values. In
simple terms:

-   ✅ **Hashable** (can be dictionary keys): strings, numbers, tuples,
    booleans
-   ❌ **Not hashable** (cannot be dictionary keys): lists,
    dictionaries, sets

Think of it this way: Python needs to quickly look up values by their
keys. To do this efficiently, the keys must be stable and unchangeable.
That’s why you can use `"name"` or `123` as keys, but not `["name"]` (a
list).

````python
user = {
    "id": 101,
    "name": "Ama",
    "skills": ["JS", "Python"]
}

# Accessing
print(user["name"]) 

# Safe Access (prevents crashing if key missing)
print(user.get("salary", 0)) # JS: user.salary || 0
````

### Building Dictionaries Dynamically

You don’t always have to create a dictionary with data already inside

````python
# Initialize an empty dictionary
student_records = {}

# Dynamically add a new key and assign it a list
student_records["names"] = ["Joshua", "Derrode"]
print(student_records)
````

## D. Sets (The “Unique List”)

Unordered collection of *unique* elements. Great for deduplicating data.

````python
raw_data = ["id_1", "id_2", "id_1", "id_3"]
unique_ids = set(raw_data) 
print(unique_ids) # {'id_3', 'id_2', 'id_1'} (Order is random!)

# Set Math (Venn Diagrams)
set_a = {1, 2, 3}
set_b = {3, 4, 5}

print(set_a.intersection(set_b)) # {3}
````

> **Real-World AI Connection: Vocabulary**
>
> We use **Sets** in data cleaning to find the **unique vocabulary** in
> a dataset. If you have 1 million tweets, you can convert the entire
> list of words into a set to instantly remove duplicates and see how
> many unique words your model needs to handle.

### Exercise 4: Set Operations

**Task:** You’re analyzing user permissions across two teams.

````python
team_a_permissions = {"read", "write", "delete", "admin"}
team_b_permissions = {"read", "write", "execute", "deploy"}

# 1. Find permissions that both teams have (intersection)
common = team_a_permissions.....(team_b_permissions)
# Alternative syntax: common = team_a_permissions ... team_b_permissions
print("Common permissions:", common)

# 2. Find all unique permissions across both teams (union)
all_permissions = team_a_permissions.....(team_b_permissions)
# Alternative syntax: all_permissions = team_a_permissions ... team_b_permissions
print("All permissions:", all_permissions)

# 3. Find permissions only Team A has (difference)
only_a = team_a_permissions.....(team_b_permissions)
# Alternative syntax: only_a = team_a_permissions ... team_b_permissions
print("Only Team A:", only_a)

# 4. Find permissions that are in either team but not both (symmetric difference)
exclusive = team_a_permissions.....(team_b_permissions)
# Alternative syntax: exclusive = team_a_permissions ... team_b_permissions
print("Exclusive permissions:", exclusive)

# 5. Add a new permission "audit" to Team A
team_a_permissions....("audit")
print("Team A after adding audit:", team_a_permissions)

# 6. Check if "admin" is in Team B permissions
is_admin_in_b = "admin" ... team_b_permissions
print("Is admin in Team B?", is_admin_in_b)
````

> **Solution**
>
> ``` python
> team_a_permissions = {"read", "write", "delete", "admin"}
> team_b_permissions = {"read", "write", "execute", "deploy"}
>
> # 1. Intersection
> common = team_a_permissions.intersection(team_b_permissions)
> # Or: common = team_a_permissions & team_b_permissions
> print("Common permissions:", common)
>
> # 2. Union
> all_permissions = team_a_permissions.union(team_b_permissions)
> # Or: all_permissions = team_a_permissions | team_b_permissions
> print("All permissions:", all_permissions)
>
> # 3. Difference (only in A)
> only_a = team_a_permissions.difference(team_b_permissions)
> # Or: only_a = team_a_permissions - team_b_permissions
> print("Only Team A:", only_a)
>
> # 4. Symmetric difference
> exclusive = team_a_permissions.symmetric_difference(team_b_permissions)
> # Or: exclusive = team_a_permissions ^ team_b_permissions
> print("Exclusive permissions:", exclusive)
>
> # 5. Add to Team A
> team_a_permissions.add("audit")
> print("Team A after adding audit:", team_a_permissions)
>
> # 6. Check membership
> is_admin_in_b = "admin" in team_b_permissions
> print("Is admin in Team B?", is_admin_in_b)
> ```
>
>     Common permissions: {'read', 'write'}
>     All permissions: {'write', 'admin', 'execute', 'read', 'delete', 'deploy'}
>     Only Team A: {'admin', 'delete'}
>     Exclusive permissions: {'delete', 'admin', 'execute', 'deploy'}
>     Team A after adding audit: {'write', 'audit', 'admin', 'read', 'delete'}
>     Is admin in Team B? False
