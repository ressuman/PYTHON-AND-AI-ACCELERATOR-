# Employee Data Cleaning & Preparation Workflow

This notebook demonstrates a **step‑by‑step data cleaning process** using pandas and numpy.

We’ll cover:
- Inspecting raw data
- Standardizing text fields
- Removing duplicates
- Handling invalid and missing values
- Cleaning numeric fields (salary, age)
- Converting dates
- Feature engineering
- Final validation

Each step includes **detailed explanations** in Markdown and corresponding Python code.

## Step 1: Create the DataFrame

```python
import pandas as pd
import numpy as np

data = {
    "employee_id": [101, 102, 103, 104, 105, 102, 106, 107, 108, 109],
    "name": [
        "Alice Johnson", "bob smith", "Charlie Brown", "Diana", "Eve Adams",
        "bob smith", "Frank ", " Grace Lee", "Hank", "Ivy Clarke"
    ],
    "department": [
        "HR", "IT", "Finance", "hr", "IT", "IT", "FINANCE", "HR", None, "Finance"
    ],
    "age": [29, 34, 45, 27, None, 34, 300, 38, 25, 31],
    "salary": [
        "$52,000", "$78,000", "$91,000", "$48,500", "$85,000",
        "$78,000", "$102,000", "$67,000", "$54,000", None
    ],
    "join_date": [
        "2021-03-15", "2019-07-22", "2018-01-10", "2022-11-05", "2020-06-30",
        "2019-07-22", "2017-09-01", "2023-02-14", "2021-08-19", "2022-04-03"
    ],
    "rating": [4.2, 3.8, 4.5, None, 4.0, 3.8, 4.1, 3.5, None, 4.3]
}

df = pd.DataFrame(data)
print(df)
```

## Step 2: Inspect the Data

```python
print(df.head())
print(df.info())
print(df.describe(include="all"))
df.describe(include="all")
```

## Step 3: Standardize Text Fields

```python
df["name"] = df["name"].str.strip().str.title()
df["department"] = df["department"].str.strip().str.title()
print(df[['name','department']])
```

## Step 4: Remove Duplicates

```python
df = df.drop_duplicates()
print(df)
```

## Step 5: Fix Invalid Ages

```python
df.loc[df["age"] > 100, "age"] = np.nan
print(df[['name','age']])
```

## Step 6: Fill Missing Ages

```python
df["age"] = df["age"].fillna(df["age"].median())
print(df[['name','age']])
```

## Step 7: Clean Salary

```python
df["salary"] = (
    df["salary"]
    .str.replace(r"[\$,]", "", regex=True)
    .astype(float)
)
print(df[['name','salary']])
```

## Step 8: Convert Join Date

```python
df["join_date"] = pd.to_datetime(df["join_date"])
print(df[['name','join_date']])
```

## Step 9: Handle Missing Ratings

```python
df["rating"] = df["rating"].fillna(df["rating"].mean())
print(df[['name','rating']])
```

## Step 10: Feature Engineering

```python
df["join_year"] = df["join_date"].dt.year
df["salary_k"] = df["salary"] / 1000
print(df[['name','join_year','salary_k']])
```

## Step 11: Final Validation

```python
print(df.head())
print(df.isnull().sum())
df.head()
df.isnull().sum()
```
