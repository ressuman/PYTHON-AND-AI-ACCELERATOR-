## Step 1: Create the DataFrame

```python
import pandas as pd
import numpy as np

exercise_df = pd.DataFrame({
    "region": ["North","South","East","West","North","South"],
    "sales": [120, 85, 90, 150, 200, 60],
    "product": ["A","B","C","A","C","B"]
})
print(exercise_df)
```

## Step 2: Inspect the Data

```python
print(exercise_df.head())
print(exercise_df.shape)
print(exercise_df.dtypes)
print(exercise_df.info())
print(exercise_df.describe())
```

## Step 3: Standardize Text Fields

```python
exercise_df["region"] = exercise_df["region"].str.strip().str.title()
exercise_df["product"] = exercise_df["product"].str.strip().str.upper()
print(exercise_df)
```

## Step 4: Remove Duplicates

```python
exercise_df = exercise_df.drop_duplicates()
print(exercise_df)
```

## Step 5: Fix Invalid Sales (simulate ages step)

```python
exercise_df.loc[exercise_df["sales"] > 1000, "sales"] = np.nan
print(exercise_df)
```

## Step 6: Fill Missing Sales with Median

```python
exercise_df["sales"] = exercise_df["sales"].fillna(exercise_df["sales"].median())
print(exercise_df)
```

## Step 7: Clean Sales (simulate salary cleaning)

```python
exercise_df["sales"] = exercise_df["sales"].astype(float)
print(exercise_df)
```

## Step 8: Add Join Date (simulate date conversion)

```python
exercise_df["join_date"] = pd.date_range(start="2021-01-01", periods=len(exercise_df), freq="MS")
print(exercise_df)
```

## Step 9: Handle Missing Ratings (simulate ratings step)

```python
exercise_df["rating"] = [4.2, 3.8, None, 4.5, 4.0, 3.9]
exercise_df["rating"] = exercise_df["rating"].fillna(exercise_df["rating"].mean())
print(exercise_df)
```

## Step 10: Feature Engineering

```python
exercise_df["join_year"] = exercise_df["join_date"].dt.year
exercise_df["sales_k"] = exercise_df["sales"] / 1000
print(exercise_df)
```

## Step 11: Final Validation

```python
print(exercise_df.head())
print(exercise_df.isnull().sum())
```

# Class Exercise Specific Tasks

### loc vs iloc

```python
print(exercise_df.loc[:, "sales"])
print(exercise_df.iloc[0, 1])
```

### Add Missing Value, Dropna, Fillna

```python
exercise_df.loc[2, "sales"] = None
print("With missing value:\n", exercise_df)
print("After dropna:\n", exercise_df.dropna())
print("After fillna:\n", exercise_df.fillna(0))
```

### GroupBy Analysis

```python
print(exercise_df.groupby("region")["sales"].mean())
```

### Method Chaining

```python
result = (
    exercise_df
    .dropna()
    .query("sales > 80")
    .groupby("region")["sales"]
    .mean()
)
print(result)
```
