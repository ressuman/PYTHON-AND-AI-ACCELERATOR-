## Example 1: String Methods

```python
# OLD (JavaScript):
# let name = "  kwame mensah  "
# name.toUpperCase()           → "  KWAME MENSAH  "
# name.trim()                  → "kwame mensah"
# name.includes("mensah")      → true
# name.replace("kwame", "ama") → "  ama mensah  "
# name.split(" ")              → ["", "", "kwame", "mensah", "", ""]

# NEW (Python):
name = "  kwame mensah  "

print(name.upper())                      # "  KWAME MENSAH  "  — JS: .toUpperCase()
print(name.strip())                      # "kwame mensah"      — JS: .trim()
print("mensah" in name)                  # True               — JS: .includes()
print(name.replace("kwame", "ama"))      # "  ama mensah  "
print(name.split())                      # ['kwame', 'mensah'] — Python auto-strips whitespace!
```

## Example 2: Conditionals

```python
# OLD (JavaScript):
# const score = 85
# if (score >= 90) {
#     console.log("A")
# } else if (score >= 75) {
#     console.log("B")
# } else {
#     console.log("C")
# }

# NEW (Python): No curly braces — indentation IS the block.
score = 85

if score >= 90:
    print("A")
elif score >= 75:      # JS: else if  →  Python: elif
    print("B")
else:
    print("C")

# Python's ternary (inline if)
# OLD JS: const label = score >= 75 ? "Pass" : "Fail"
label = "Pass" if score >= 75 else "Fail"
print(label)  # Pass
```

## Example 3: Loops

```python
# OLD (JavaScript):
# const models = ["GPT-4", "Gemini", "Claude"]
#
# // for...of loop
# for (const model of models) {
#     console.log(model)
# }
#
# // forEach with index
# models.forEach((model, index) => {
#     console.log(index, model)
# })
#
# // Classic for loop
# for (let i = 0; i < 5; i++) {
#     console.log(i)
# }

# NEW (Python):
models = ["GPT-4", "Gemini", "Claude"]

# for...of equivalent
for model in models:
    print(model)

print("---")

# forEach with index → Python uses enumerate()
for index, model in enumerate(models):
    print(index, model)

print("---")

# Classic for loop → Python uses range()
for i in range(5):
    print(i)

print("---")

# Bonus: loop with a start index (no clean equivalent in JS forEach)
for index, model in enumerate(models, start=1):
    print(f"{index}. {model}")
```

## Example 4: Functions

```python
# OLD (JavaScript):
# // Regular function
# function greet(name, role = "Developer") {
#     return `Hello, ${name}! You are a ${role}.`
# }
#
# // Arrow function
# const double = (n) => n * 2

# NEW (Python):
# Regular function — "def" replaces "function", no curly braces
def greet(name, role="Developer"):     # default parameter works the same
    return f"Hello, {name}! You are a {role}."

print(greet("Ama"))
print(greet("Kwame", "AI Engineer"))

# Arrow function equivalent → Python uses "lambda"
double = lambda n: n * 2
print(double(5))  # 10

# Pythonic note: for anything longer than one expression, use def, not lambda
```

## Example 5: List Comprehensions (No Direct JS Equivalent)

```python
# OLD (JavaScript):
# // You'd use .map() and .filter()
# const scores = [45, 72, 88, 91, 60]
#
# const passing      = scores.filter(s => s >= 70)
# const doubled      = scores.map(s => s * 2)
# const passAndDouble = scores.filter(s => s >= 70).map(s => s * 2)

# NEW (Python):
# Python has a built-in syntax for this — it's called a list comprehension.
# Pattern: [expression for item in list if condition]

scores = [45, 72, 88, 91, 60]

passing       = [s for s in scores if s >= 70]           # JS: .filter()
doubled       = [s * 2 for s in scores]                   # JS: .map()
pass_doubled  = [s * 2 for s in scores if s >= 70]        # JS: .filter().map()

print("Passing:", passing)        # [72, 88, 91]
print("Doubled:", doubled)        # [90, 144, 176, 182, 120]
print("Pass & Doubled:", pass_doubled)  # [144, 176, 182]
```

## Example 6: Spreading / Unpacking

```python
# OLD (JavaScript):
# const a = [1, 2, 3]
# const b = [4, 5, 6]
# const combined = [...a, ...b]          // [1,2,3,4,5,6]
#
# const obj1 = { name: "Ama" }
# const obj2 = { role: "Engineer" }
# const merged = { ...obj1, ...obj2 }    // { name: "Ama", role: "Engineer" }

# NEW (Python):
a = [1, 2, 3]
b = [4, 5, 6]

# Spread lists → use the * operator
combined = [*a, *b]
print(combined)     # [1, 2, 3, 4, 5, 6]

# Or simply use +
combined_v2 = a + b
print(combined_v2)  # [1, 2, 3, 4, 5, 6]

# Spread / merge dicts → use the ** operator
obj1 = {"name": "Ama"}
obj2 = {"role": "Engineer"}
merged = {**obj1, **obj2}
print(merged)       # {'name': 'Ama', 'role': 'Engineer'}

# Unpacking in function calls (same as JS spread in function args)
def add(x, y, z):
    return x + y + z

nums = [1, 2, 3]
print(add(*nums))   # 6  — JS: add(...nums)
```

## Example 7: Error Handling

```python
# OLD (JavaScript):
# try {
#     const result = riskyOperation()
# } catch (error) {
#     console.error("Something went wrong:", error.message)
# } finally {
#     console.log("Always runs")
# }

# NEW (Python):
# "catch" becomes "except", everything else maps 1:1
def risky_operation(value):
    if value == 0:
        raise ValueError("Cannot use zero!")   # JS: throw new Error(...)
    return 100 / value

try:
    result = risky_operation(0)
except ValueError as error:           # catch specific error type
    print(f"Caught a ValueError: {error}")
except ZeroDivisionError as error:    # you can chain multiple except blocks
    print(f"Caught a ZeroDivisionError: {error}")
except Exception as error:            # catch-all (like catch(error) in JS)
    print(f"Caught unknown error: {error}")
finally:
    print("This always runs — like JS finally")
```

## Example 8: Classes & OOP

```python
# OLD (JavaScript):
# class AIModel {
#     constructor(name, version) {
#         this.name = name
#         this.version = version
#     }
#     describe() {
#         return `${this.name} v${this.version}`
#     }
# }
# const model = new AIModel("Claude", 3)
# console.log(model.describe())

# NEW (Python):
# "constructor" → "__init__"
# "this"        → "self"   (must be the first parameter in every method)
# No "new" keyword needed

class AIModel:
    def __init__(self, name: str, version: float):  # constructor
        self.name = name          # "this.name" in JS
        self.version = version

    def describe(self):           # every method gets "self" as first param
        return f"{self.name} v{self.version}"

    # Python "magic method" — like toString() in JS
    def __str__(self):
        return self.describe()

model = AIModel("Claude", 3.5)    # No "new" keyword!
print(model.describe())           # Claude v3.5
print(model)                      # Claude v3.5  ← calls __str__ automatically
```

## Example 9: Working with JSON / Dicts (API Responses)

```python
# OLD (JavaScript):
# const jsonStr = '{"user": "Ama", "score": 95}'
# const obj = JSON.parse(jsonStr)       // string → object
# const backToStr = JSON.stringify(obj) // object → string
# console.log(obj.user)                 // "Ama"

# NEW (Python):
import json  # You need to import the json module (like require() in Node.js)

json_str = '{"user": "Ama", "score": 95}'

obj = json.loads(json_str)          # JSON.parse()      →  json.loads()
print(obj["user"])                  # "Ama" — note: dict uses ["key"], not .key
print(type(obj))                    # <class 'dict'>

back_to_str = json.dumps(obj)       # JSON.stringify()  →  json.dumps()
print(back_to_str)

# Pretty print (like JSON.stringify(obj, null, 2))
pretty = json.dumps(obj, indent=2)
print(pretty)
```

## Example 10: Importing Modules

```python
# OLD (JavaScript / Node.js):
# const fs = require('fs')                     // CommonJS
# import { readFile } from 'fs'               // ES Module named import
# import express from 'express'               // ES Module default import
# import { useState, useEffect } from 'react'  // multiple named imports

# NEW (Python):
import math                          # like: import math from 'math'

# Named imports — "from" goes FIRST in Python (opposite of JS!)
from math import sqrt, pi            # like: import { sqrt, pi } from 'math'

# Import with alias — same idea as JS "as"
# import numpy as np                 # like: import numpy as np (if installed)
# from datetime import datetime as dt

print(math.sqrt(144))   # 12.0  — full module reference
print(sqrt(144))         # 12.0  — direct reference after named import
print(pi)                # 3.14159...

# Key rule:
# "from X import Y"  =  named import   → JS: import { Y } from 'X'
# "import X"         =  default import → JS: import X from 'X' (access via X.something)
```
